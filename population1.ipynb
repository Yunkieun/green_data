```python
import csv
import matplotlib.pyplot as plt

with open('datas/age_2023.csv', 'r') as f:
    data = csv.reader(f)
    next(data)
    result = []  # 연령별 인구 저장할 리스트
    # 지역이름 입력받기
    name = input("지역의 이름(읍면동 단위)을 입력해 주세요:")

    for row in data:
        if name in row[0]:
            for i in row[3:]:  #3번 인덱스 ~ 104번(104~1) 인덱스까지 반복
                result.append(int(i.replace(",", '')))  #  천단위 구부기호 제거
                
print(len(row[3:]))
#print(result)
plt.rc('font', family="Malgun Gothic")
plt.style.use("ggplot")  # 격자 무늬 스타일
plt.title(name + "지역의 인구 구조")
plt.plot(result, 'r')  # result - y축값
# plt.bar(range(101), result)  # 막대그래프 : x축-(0~100), y축-result
plt.xlabel('연령')
plt.ylabel('인구')
plt.show()
```
